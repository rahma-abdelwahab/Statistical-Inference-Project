```r
 #Function to detect outliers in all columns
detect_outliers <- function(data) {
  outlier_info <- list()  # To store the outliers

  # Iterate over all columns to calculate the bounds
  for (col in names(data)) {
    Q1 <- quantile(data[[col]], 0.25)
    Q3 <- quantile(data[[col]], 0.75)
    IQR <- Q3 - Q1

    # Calculate lower and higher bounds
    lower_bound <- Q1 - 1.5 * IQR
    upper_bound <- Q3 + 1.5 * IQR

    # Identify outliers in the column
    outliers <- data[[col]][data[[col]] < lower_bound | data[[col]] > upper_bound]

    # Store the outliers in the list if any
    if (length(outliers) > 0) {
      outlier_info[[col]] <- outliers
    }
  }

  return(outlier_info)
}

# Check for outliers in the dataset
outliers <- detect_outliers(data)

# Print out the outliers if found
if (length(outliers) > 0) {
  cat("Outliers detected in the following columns:\n")
  print(outliers)
} else {
  cat("No outliers detected in the dataset.\n")
}
```

```r
#After detecteing the outlier, we need to remove them from the data
# Function to remove outliers in all columns
remove_outliers <- function(data, outlier_info) {
  # Iterate over all columns to filter out rows with outliers
  for (col in names(data)) {
    # Calculate the bounds from the outlier_info
    if (col %in% names(outlier_info)) {
      lower_bound <- quantile(data[[col]], 0.25) - 1.5 * (quantile(data[[col]], 0.75) - quantile(data[[col]], 0.25))
      upper_bound <- quantile(data[[col]], 0.75) + 1.5 * (quantile(data[[col]], 0.75) - quantile(data[[col]], 0.25))

      # Remove rows that have outliers in the current column
      data <- data[data[[col]] >= lower_bound & data[[col]] <= upper_bound, ]
    }
  }

  return(data)
}

# Remove the outliers
data <- remove_outliers(data, outliers)

# Print the cleaned data (after outliers are removed)
print(head(data))

```

## 3.1 Claim: “There is a significant difference in glucose levels between diabetic and non-diabetic patients.”

### **step 1: State the hypothyses**
Null Hypothesis ($H_{0}$): There is no significant difference in glucose levels between diabetic and non-diabetic patients.

$H_{0}:μ_{diabetic}=μ_{non-diabetic}$

Alternative Hypothesis ($H_{α}$): There is a significant difference in glucose levels between diabetic and non-diabetic patients.

$H_{α}: μ_{diabetic} \neq μ_{non-diabetic}$

### **step 2: Select the Test**
The approprtiate test here is two-tailed t-test because:
* Assuming that glucose levels are approximately normally distributed in each group.
* Assuming that the variances of the two groups are equal.
* The sample size is large enough for the central limit theorem to apply.


### **Step 3: Conduct the test**

```r
t_test_claim_1 <- t.test(Glucose ~ Outcome, data = data, var.equal = TRUE)

t_test_claim_1
```

###**step 4: Test Statistics and P-value**

The t-test has shown:

* t-value
* degrees of freedom (df)
* p-value

By comparing the p-value with the significance level(α = 0.05):

the p-value < 0.05: Reject the null hypothesis.

There is a significant difference in glucose levels between diabetic and non-diabetic patients.

###**step5: conclusion**

With a p-value of 2.2e-16, we will reject the null hypothesis (H₀) and conclude that there is a statistically significant difference in glucose levels between diabetic and non-diabetic patients.
The evidence strongly supports the alternative hypothesis (Hₐ) that there is a significant difference in glucose levels between the two groups.
